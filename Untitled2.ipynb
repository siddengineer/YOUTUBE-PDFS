```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Step 1: Raw Data
states = [
    "Himachal Pradesh", "Goa", "Jammu-Kashmir", "Sikkim", "Meghalaya",
    "Arunachal Pradesh", "Tripura", "Nagaland", "Manipur", "Mizoram"
]

gst_values = [1022, 449, 666, 364, 172, 88, 57, 54, 40, 30]

# Step 2: Feature Engineering
df = pd.DataFrame({
    'State': states,
    'GST': gst_values
})
df['GST_Rank'] = df['GST'].rank(ascending=False)
df['GST_Percent'] = df['GST'] / df['GST'].max()
df['GST_Log'] = np.log1p(df['GST'])  # log(1+gst) for stability

# Step 3: Encode Categorical (State) to Integer
le = LabelEncoder()
df['State_ID'] = le.fit_transform(df['State'])

# Normalize continuous features
scaler = MinMaxScaler()
features = scaler.fit_transform(df[['GST_Rank', 'GST_Percent', 'GST_Log']])

# Step 4: Build TensorFlow Model with Embedding
input_state = tf.keras.Input(shape=(1,), name='state_input')
state_embed = tf.keras.layers.Embedding(input_dim=len(states), output_dim=4)(input_state)
state_embed_flat = tf.keras.layers.Reshape((4,))(state_embed)

input_numeric = tf.keras.Input(shape=(3,), name='features_input')

x = tf.keras.layers.Concatenate()([state_embed_flat, input_numeric])
x = tf.keras.layers.Dense(32, activation='relu')(x)
x = tf.keras.layers.Dense(16, activation='relu')(x)
output = tf.keras.layers.Dense(1, name='gst_output')(x)

model = tf.keras.Model(inputs=[input_state, input_numeric], outputs=output)
model.compile(optimizer='adam', loss='mse')
model.summary()

# Step 5: Train the model
model.fit(
    {'state_input': df['State_ID'], 'features_input': features},
    df['GST'],
    epochs=500,
    verbose=0
)

# Step 6: Extract State Embeddings
embedding_matrix = model.get_layer('embedding').get_weights()[0]
print("\n📌 Learned State Embeddings:\n", embedding_matrix)

# Step 7: Cosine Similarity
similarity = cosine_similarity(embedding_matrix)
similarity_df = pd.DataFrame(similarity, index=states, columns=states)
print("\n🔍 Cosine Similarity Between States:\n")
print(similarity_df.round(2))

# Step 8: Visualize Embeddings using t-SNE
tsne = TSNE(n_components=2, perplexity=5, random_state=42)
emb_2d = tsne.fit_transform(embedding_matrix)

plt.figure(figsize=(8,6))
for i, state in enumerate(states):
    plt.scatter(emb_2d[i, 0], emb_2d[i, 1])
    plt.text(emb_2d[i, 0]+0.01, emb_2d[i, 1]+0.01, state, fontsize=9)
plt.title('State Embeddings Visualized via t-SNE')
plt.xlabel("Dim 1")
plt.ylabel("Dim 2")
plt.grid(True)
plt.show()
```
